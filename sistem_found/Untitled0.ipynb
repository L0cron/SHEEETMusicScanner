```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow  # Импортируйте cv2_imshow
```

```python
main_image = cv2.imread('/content/drive/MyDrive/ds/mppo/123.png')
template = cv2.imread('/content/drive/MyDrive/ds/mppo/photo_2024-12-11_18-29-00.jpg')
```

```python
main_gray = cv2.cvtColor(main_image, cv2.COLOR_BGR2GRAY)
template_gray = cv2.cvtColor(template, cv2.COLOR_BGR2GRAY)
```

```python
# Шаблонное сопоставление
result = cv2.matchTemplate(main_gray, template_gray, cv2.TM_CCOEFF_NORMED)
threshold = 0.65  # Установите порог для совпадения
yloc, xloc = np.where(result >= threshold)

# Отрисовка прямоугольников вокруг найденных объектов
for (x, y) in zip(xloc, yloc):
    cv2.rectangle(main_image, (x, y), (x + template.shape[1], y + template.shape[0]), (0, 255, 0), 2)

# Показать результат
cv2_imshow(main_image)
```
